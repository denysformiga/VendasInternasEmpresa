# Criando uma nova coluna para as metas

```python
# Duplicando o dataset para não alterar o original
Meta = dim_agents
```

```python
Meta.head()
```

```python
# Convertendo a coluna Nível para float, afinal, é um número e não uma categoria
Meta['nivel'] = Meta['nivel'].astype('float64')
```

```python
# Transformando os dados de acordo com as regras de negócio

meta_por_nivel = {1: 1, 2: 1.1, 3: 1.2, 4: 1.25, 5: 1.3}
Meta['nivel'] = Meta['nivel'].map(meta_por_nivel)

meta_por_canal = {'chat': 3, 'whatsapp': 4, 'telefone': 3.7}
Meta['canal'] = Meta['canal'].map(meta_por_canal)
```

```python
Meta.head()
```

```python
Meta.info()
```

```python
Meta['canal'] = Meta['canal'].astype('float64')
```

```python
Meta['meta_diaria'] = Meta['nivel'] * Meta['canal']
```

```python
Meta['meta_diaria'] = Meta['meta_diaria'].round(2)
```

```python
Meta.head()
```

```python
Meta.info()
```

```python
# Agrupando a quantidade de vendas por Agent_ID e por Data
```

```python
fact_sales.head()
```

```python
resultado = fact_sales.groupby(['Agent_ID', 'Date'],  observed=False).size().reset_index(name='quantidade_vendas')

# Renomeie a coluna de contagem para algo mais descritivo
fact_sales_sum = resultado.rename(columns={'sale_id': 'quantidade_vendas'})

fact_sales_sum.shape
```

```python
# Curiosamente o shape do ultimo DF deu 18000 (600 * 30) e não 18600 (600 * 31). Vamos verificar as datas únicas.
a = fact_sales_sum.Date.unique()
print(*a, sep='\n')
```

### Curiosamente não houve nenhuma venda no dia 24 (quarto domingo do mês). 
### Este resultado também foi obtido no MS Excel utilizando o arquivo 'fact_sales.csv' original.
### Esta informação coincide com a questão de que "Para agentes que trabalham no domingo, há uma regra em que após três domingos trabalhados, 
### o colaborador deve ter um domingo de folga. Para isso, considere que o primeiro domingo trabalhado do time será dia 03/01/2077."
### Logo, não há o que implementar uma função para os domingos, dado que não há registros de Dezembro de 2076.

# Juntando as tabelas

```python
# Agora podemos realizar a junção das tabelas

colunas_df1 = ['Agent_ID', 'Date', 'quantidade_vendas']
colunas_df2 = ['Agent_ID', 'folga_1', 'folga_2', 'meta_diaria'] 

resultado = pd.merge(fact_sales_sum[colunas_df1], Meta[colunas_df2], on='Agent_ID')
```

```python
resultado.head()
```

```python
resultado.tail()
```

```python
resultado.info()
```

```python
resultado['Date'] = pd.to_datetime(resultado['Date'])

# Extrair o nome do dia da semana e atribuir a uma nova coluna
resultado['weekday'] = resultado['Date'].dt.day_name()
```

```python
resultado.head()
```

```python
# Convertendo os dias da semana em números para melhor manipulação
dia_por_numero = {'Sunday': 1, 'Monday': 2, 'Tuesday': 3, 'Wednesday': 4, 'Thursday': 5,'Friday': 6, 'Saturday': 7}
resultado['weekday'] = resultado['weekday'].map(dia_por_numero)
```

```python
resultado.head()
```

```python
# Convertendo as colunas para tipo inteiro
resultado[['weekday', 'folga_1', 'folga_2']] = resultado[['weekday', 'folga_1', 'folga_2']].astype('int64')
```

```python
resultado.info()
```

```python
# Identificando dias de trabalho e folga de acordo com a coluna folga1 e folga2
resultado['trabalha'] = ~((resultado['folga_1'] == resultado['weekday']) | (resultado['folga_2'] == resultado['weekday']))
```

```python
resultado.head()
```

```python
# Adicionando o feriado do dia primeiro, para não contabilizar a meta do agente neste dia
resultado.loc[resultado['Date'] == '2077-01-01', 'trabalha'] = False
```

```python
resultado.head()
```

```python
# Entrega da questão de número 1 do case 

resultado.to_csv('empresa1.csv', index=False)
```
